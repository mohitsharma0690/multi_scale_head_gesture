```python
### Plot pose velocity values 

def get_bin_count_for_gesture(X, y, gest, bins):
    bin_hist = [0]*(len(bins)+1)
    for i in range(X.shape[0]):
        if data_utils.get_classif_class_for_gest_5(y[i]) == gest:
            idx = data_utils.get_bin_in_sorted_list(bins, X[i])
            bin_hist[idx] += 1
    return bin_hist

def filter_gesture(X, y, gest):
    X_filt, idx = np.zeros(X.shape), 0
    for i in range(X.shape[0]):
        if data_utils.get_classif_class_for_gest_5(y[i]) == gest:
            if len(X_filt.shape) == 1:
                X_filt[idx] = X[i]
            else:
                X_filt[idx,:] = X[i,:]
            idx = idx + 1
    
    return X_filt[:idx, :] if len(X_filt.shape) > 1 else X_filt[:idx]

def plot_pose_velocity(X, y, curr_file):
    X_pose_vel = X[:,6:9]
    for i in range(X_pose_vel.shape[1]):
        X_pose_vel[:, i], _ = data_utils.reject_outliers(
            X_pose_vel[:,i], m=20.)
    X_pose_vel_diff = X_pose_vel[1:,:] - X_pose_vel[:-1,:]
    print("Pose velocity distribution {}", curr_file)
    for i in range(X_pose_vel_diff.shape[1]):
        if i != 0: continue
        X1 = X_pose_vel_diff[:,i]
        print("MEAN: {}, STD: {}, MEDIAN: {}".format(
            np.mean(X1), np.std(X1), np.median(X1)))
        bins = [0.001]
        for i in range(20):
            bins.append(bins[-1]+0.1)
        
        X1 = np.abs(X1)

        # X1 = (X1- np.mean(X1)) / np.std(X1)
        none_hist = get_bin_count_for_gesture(X1, y, 0, bins)
        nod_hist = get_bin_count_for_gesture(X1, y, 1, bins)
        none_filt = filter_gesture(X1, y, 0)
        nod_filt = filter_gesture(X1, y, 1)
        tilt_filt = filter_gesture(X1, y, 2)
        shake_filt = filter_gesture(X1, y, 3)
        fwd_filt = filter_gesture(X1, y, 4)
        
        X1, vel_outliers = data_utils.reject_outliers(X1, m=20.)
        if vel_outliers > 0:
            print("Velocity outliers count: {}".format(vel_outliers))
            
        print("Axis {}", i)

        # bins = range(0.01, 0.1, 0.05)
        hist, bin_edges = np.histogram(X1, bins=bins)
        print("# hist {}".format(np.sum(hist)))
        plt.hist(X1, bins=bins)
        plt.show()
        plt.hist([none_filt, nod_filt, tilt_filt, shake_filt, fwd_filt],
                 bins=bins, label=['None', 'Nod', 'Tilt', 'Shake', 'Fwd'],
                )
        plt.legend()
        plt.show()
        ax = sns.distplot(X1);
        plt.show()
        plt.plot(range(len(X1[200:])), X1[200:])
        plt.show()
        

def read_h5_dir(h5_dir, call_func=None):
    for f in sorted(os.listdir(h5_dir)):
        if f.endswith('static.mp4.txt.h5'):
            h5_f = h5py.File(os.path.join(h5_dir, f), 'r')
            X = np.array(h5_f['features'])
            y = np.array(h5_f['annotations'])
            if call_func is not None:
                call_func(X, y, f)
                # plot_pose_velocity(X, y, f)
            
PLOT_DIST = 1
if PLOT_DIST:
    h5_dir = '../openface_data/gest_data_38_2/correct_gests_38/'
    read_h5_dir(h5_dir, call_func=plot_pose_velocity)
```
